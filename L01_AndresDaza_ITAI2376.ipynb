```python
# Import necessary libraries
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
import io
import ipywidgets as widgets
```

# Introduction
This lab is designed to introduce you to the basics of deep learning by interacting with a pre-built model. You'll understand the workflow of a deep learning project, including data preprocessing, model architecture, and making predictions. The goal is to familiarize yourself with the basics of deep learning without writing any code.


```python
# Load the VGG16 model
model = VGG16(weights='imagenet')

# Display the model architecture
model.summary()
```

```python
# Load and preprocess an image
def load_and_preprocess_image(image_path):
    # Load the image
    img = load_img(image_path, target_size=(224, 224))

    # Convert the image to a numpy array
    img_array = img_to_array(img)

    # Expand dimensions to fit the model input
    img_array = np.expand_dims(img_array, axis=0)

    # Preprocess the image
    img_array = preprocess_input(img_array)

    return img, img_array

# Load and preprocess a sample image
sample_image, processed_image = load_and_preprocess_image('cat.png')

# Display the sample image
plt.imshow(sample_image)
plt.show()
```

```python
# Make predictions
predictions = model.predict(processed_image)

# Decode and print the predictions
decoded_predictions = decode_predictions(predictions, top=3)[0]
print(decoded_predictions)
```

```python
# Upload button to load images
upload = widgets.FileUpload()
display(upload)

# Button to make predictions
predict_button = widgets.Button(description="Make Prediction")
display(predict_button)

# Function to handle button click
def on_click(change):
    img_data = list(upload.value.values())[0]['content']
    img = Image.open(io.BytesIO(img_data))
    img = img.convert('RGB') # Convert image to RGB to ensure 3 channels
    img = img.resize((224, 224))

    # Preprocess and predict
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    predictions = model.predict(img_array)
    decoded_predictions = decode_predictions(predictions, top=3)[0]

    # Display predictions
    print(decoded_predictions)

predict_button.on_click(on_click)
```

##R=
Pre-trained models such as VGG16 generate predictions by identifying patterns learned from extensive training on large-scale datasets, such as ImageNet. Through this training process, the model learns hierarchical features, ranging from simple edges and textures to complex object components, that enable it to recognize and classify objects effectively.
Data preprocessing plays a important role in ensuring that input images match the format the model was trained on. This typically involves tasks such as resizing images to a fixed resolution (e.g., 224x224 pixels), normalizing pixel values, and ensuring the correct number of color channels. Without proper preprocessing, the input data may fall outside the distribution the model expects, leading to unreliable or inaccurate predictions.
Similarly, any modifications to the input, such as altering image size, changing color channels, applying filters, or adjusting brightness or contrast—can significantly influence the model's output. When these changes distort or remove key visual features that the model relies on, performance may degrade or predictions may become incorrect. Conversely, when preprocessing is performed appropriately and consistently, it preserves the essential features the model was trained to recognize, allowing it to make accurate and dependable predictions.
